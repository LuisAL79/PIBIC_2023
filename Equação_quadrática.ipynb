```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def func(x1,f1,x2,f2,x3,f3):
  y1 = a*x1**2 + b*x1 + c
  eq1 = Eq(y1,f1)
  y2 = a*x2**2 + b*x2 + c
  eq2 = Eq(y2,f2)
  y3 = a*x3**2 + b*x3 + c
  eq3 = Eq(y3,f3)
  sol = solve([eq1,eq2,eq3],[a,b,c])
  fsol = sol[a]*x**2 + sol[b]*x + sol[c]

  return fsol

a,b,c, x = symbols('a b c x')
func(-1,-4,1,2,2,-1)
```

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def real(expr,u,v):
  a, b, c = Poly(expr.lhs, u).all_coeffs()
  op = expr.rel_op ; dis = b**2 - 4*a*c

  if a.has(v):
    if op == '<':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a<0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
    elif op == '<=':
      sol = solveset(dis<=0,v, domain=S.Reals)
      sola = solveset(a<=0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
    elif op == '>':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a>0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
    elif op == '>=':
      sol = solveset(dis<=0,v, domain=S.Reals)
      sola = solveset(a>=0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
  else:
    if op == '<' and a < 0 :
      solu = solveset(dis<0,v, domain=S.Reals)
      return solu
    elif op == '<=' and a < 0:
      solu = solveset(dis<=0,v, domain=S.Reals)
      return solu
    elif op == '>' and a > 0:
      solu = solveset(dis<0,v, domain=S.Reals)
      return solu
    elif op == '>=' and a >0:
      solu = solveset(dis<=0,v, domain=S.Reals)
      return solu
    else:
      return f'Vazio'

x, m = symbols('x m')
exp = (m**2-1)*x**2 + 2*(m-1)*x + (1) > 0
real(exp,x,m)
```

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def solfrac(ineq,u,v):
  ineq = ineq.simplify() ; op = ineq.rel_op
  nume, den = ineq.lhs.as_numer_denom()
  a, b, c = Poly(nume, u).all_coeffs()
  dis = b**2 - 4*a*c

  if a.is_number:
    if op == '<' and a < 0 :
      solu = solveset(dis<0,v, domain=S.Reals)
      return ineq,solu
    elif op == '>' and a > 0:
      solu = solveset(dis<0,v, domain=S.Reals)
      return ineq,solu
  elif a.has(v):
    if op == '<':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a<0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return ineq,solu
    elif op == '>':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a>0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return ineq,solu
  else:
      return f'Vazio'

m,x = symbols('m x')

expr = (x)/(x**2 + 4) - (x+m )/(x**2+1) > 0
solfrac(expr,x,m)
```
