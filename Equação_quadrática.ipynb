```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def solve_for_m(equation, condition, m, x1, x2):
    coeffs = equation.as_poly(x).all_coeffs()
    deg = len(coeffs) - 1

    # Relações de Viète para equações de qualquer grau
    roots = symbols(f'x1:{deg+1}')
    viete_relations = {}
    # Soma das raízes: -a_{n-1}/a_n
    sum_roots = -coeffs[1] / coeffs[0]
    viete_relations[sum(roots)] = sum_roots
    # Produto das raízes: (-1)^n * a_0 / a_n
    prod_roots = Mul(*roots)  # Multiplicar todas as raízes
    viete_relations[prod_roots] = (-1)**deg * coeffs[-1] / coeffs[0]
    # Outras somas de produtos dois a dois, três a três, etc.
    for i in range(2, deg + 1):
        sum_prod_roots = sum([roots[j] * roots[k] for j in range(deg) for k in range(j + 1, deg)])
        viete_relations[sum_prod_roots] = coeffs[i] / coeffs[0]
    s,p = viete_relations.values()
    # Substituir as relações de Viète na condição
    num, deno = together(condition).lhs.as_numer_denom()
    n = Poly(num).degree(x1)

    def sum_of_powers(s, p, n):
      if n == 0:
        return 2  # x1^0 + x2^0 = 1 + 1 = 2
      elif n == 1:
        return s  # x1 + x2 = s
      elif n == 2:
        return s**2 - 2*p  # x1^2 + x2^2
      else:
        return s * sum_of_powers(s, p, n-1) - p * sum_of_powers(s, p, n-2)

    res=sum_of_powers(s,p,n)/deno.subs({x1*x2:p})
    sol = solve(Eq(res,condition.rhs), m)

    return sol

m, x, x1, x2 = symbols('m x x1 x2')
equation = m*x**2 - 2*(m-1)*x + m
condition = Eq(x1/x2 + x2/x1, 4)
solve_for_m(equation, condition, m, x1, x2)
```

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def value_m(expr, value, u, v, cond):
  a, b, c = Poly(expr, u).all_coeffs()
  yv = (4 * a * c - b**2) / (4 * a)  # Valor do vértice
  m_value = solve(Eq(yv, value), v)  # Solução para m

  if a.is_number:
      if (cond == 'Min' and a > 0) or (cond == 'Max' and a < 0):
            return m_value
  else:
      u_ineq = solveset((a > 0) if cond == 'Min' else (a < 0), v, domain=S.Reals)
      return Intersection(FiniteSet(*m_value), u_ineq)

x, m = symbols('x m')
fx = (m-1)*x**2 + (m+1)*x - m
val = 1
result = value_m(fx, val, x, m, 'Min')
result
```

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def real(expr,u,v):
  a, b, c = Poly(expr.lhs, u).all_coeffs()
  op = expr.rel_op ; dis = b**2 - 4*a*c

  if a.has(v):
    if op == '<':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a<0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
    elif op == '<=':
      sol = solveset(dis<=0,v, domain=S.Reals)
      sola = solveset(a<=0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
    elif op == '>':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a>0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
    elif op == '>=':
      sol = solveset(dis<=0,v, domain=S.Reals)
      sola = solveset(a>=0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return solu
  else:
    if op == '<' and a < 0 :
      solu = solveset(dis<0,v, domain=S.Reals)
      return solu
    elif op == '<=' and a < 0:
      solu = solveset(dis<=0,v, domain=S.Reals)
      return solu
    elif op == '>' and a > 0:
      solu = solveset(dis<0,v, domain=S.Reals)
      return solu
    elif op == '>=' and a >0:
      solu = solveset(dis<=0,v, domain=S.Reals)
      return solu
    else:
      return f'Vazio'

x, m = symbols('x m')
exp = (m**2-1)*x**2 + 2*(m-1)*x + (1) > 0
real(exp,x,m)
```

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def solfrac(ineq,u,v):
  ineq = ineq.simplify() ; op = ineq.rel_op
  nume, den = ineq.lhs.as_numer_denom()
  a, b, c = Poly(nume, u).all_coeffs()
  dis = b**2 - 4*a*c

  if a.is_number:
    if op == '<' and a < 0 :
      solu = solveset(dis<0,v, domain=S.Reals)
      return ineq,solu
    elif op == '>' and a > 0:
      solu = solveset(dis<0,v, domain=S.Reals)
      return ineq,solu
  elif a.has(v):
    if op == '<':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a<0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return ineq,solu
    elif op == '>':
      sol = solveset(dis<0,v, domain=S.Reals)
      sola = solveset(a>0,v, domain=S.Reals)
      solu = Intersection(sol,sola)
      return ineq,solu
  else:
      return f'Vazio'

m,x = symbols('m x')

expr =  x/(x**2+4) - ((x+m )/(x**2+1)) > 0
solfrac(expr,x,m)
```

```python
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def raiz3(expr, low, sup, u, v):
  a, b, c = Poly(expr, u).all_coeffs()
  discriminant = b**2 - 4*a*c
  vertex = -b / (2*a)
  # Condição para os limites inferior e superior
  cond_inf = solveset(a * expr.subs(u, low) > 0, v, domain=S.Reals)
  cond_sup = solveset(a * expr.subs(u, sup) > 0, v, domain=S.Reals)
  soldisc = solveset(discriminant > 0, v, domain=S.Reals)
  # Verificar se o vértice está entre low e sup
  sol_low = solveset(vertex > low, v, domain=S.Reals)
  sol_sup = solveset(vertex < sup, v, domain=S.Reals)

  sol = Intersection(soldisc, cond_inf, sol_low, cond_sup, sol_sup)
  return sol

x, m = symbols('x m')
fx = (m)*x**2 - 2*(m+1)*x+(m+5)
inf = 0 ; sup = 2
raiz3(fx, inf, sup, x, m)
```
