```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

def resoluo(exp1,u,t):
  gf1 = lambdify(u,exp1,'numpy')
  m=exp1.coeff(u)
  yu_latex=latex(exp1)

  plt.plot(t,gf1(t),label=f'$y={yu_latex}$')
  if m > 0:
    plt.title('funao crescente a>0')
  else:
    plt.title('funao decrescente a<0')

  plt.legend()
  plt.grid()
  plt.show()

  return

x = symbols('x')
exp1 = 2*x-1
t= np.linspace(-10,10)
resoluo(exp1,x,t)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def completa(exp1,exp2,u,v,t):
  eq1=Eq(exp1,0) ; eq2=Eq(exp2,0)
  sol = solve([eq1,eq2],[u,v])

  f1 = solve(eq1.lhs,y)[0]
  f2 = solve(eq2.lhs,y)[0]
  y1=latex(f1) ; y2=latex(f2)
  gf1 = lambdify(u,f1,'numpy')
  gf2 = lambdify(u,f2,'numpy')

  plt.plot(t,gf1(t),label=f'$y={y1}$')
  plt.plot(t,gf2(t),label=f'$y={y2}$')
  plt.plot(sol[u],sol[v],'o',label='soluo',color='black')

  plt.legend()
  plt.grid()
  print('soluo do sistema')

  return sol

x,y = symbols('x y')
exp1 = x+y-5
exp2 = x-y-1
t= np.linspace(-9,9)

completa(exp1,exp2,x,y,t)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

def reta(plx,ply,p2x,p2y,u,t):
  eq = Eq( (p2y-ply)/(p2x-plx) , (y - p2y)/(x-p2x) )
  sol = solve(eq,y)
  exp1 = sol[0]
  gf1 = lambdify(u,exp1,'numpy')
  m=exp1.coeff(u)
  yu_latex=latex(exp1)

  plt.plot(t,gf1(t),label=f'$y={yu_latex}$')
  if m > 0:
    plt.title('funao crescente a>0')
  else:
    plt.title('funao decrescente a<0')
  plt.grid()
  plt.legend()

  return

x = symbols('x')
t = np.linspace(-8,8)
reta(1,-1,-1,2,x,t)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

def lin_graph(b,x_0,y_0,u,t):
  a = Rational(b-y_0)/Rational(-x_0)
  yu=a*u+b
  yu_latex=latex(yu)

  fx = lambdify(u,yu,'numpy')
  plt.plot(t,fx(t),label=f'$y={yu_latex}$')

  if a > 0 :
    plt.title('funao crescente, a>0')
  else:
    plt.title('funao decrescente, a<0')
  plt.legend(fontsize=14)
  plt.grid()
  plt.show()

  return

x = symbols('x')
intervalo = np.linspace(0,12)
lin_graph(-3,-3,-2,x,intervalo)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy.interactive import printing
printing.init_printing(use_latex=True)

def variacao(exp,u,v,t):
  y = symbols('y')
  exp = exp.expand().factor(u)
  a=exp.coeff(u)
  gf1 = lambdify((u,v),exp,'numpy')
  p = solve(a,v)[0]

  yco=latex(exp.subs(v,p))
  yc=latex(exp.subs(v,p+1))
  yd=latex(exp.subs(v,p-1))
  pol = Eq(y,exp)

  plt.title(f'Crescente: ${v>p}$, Decrescente: ${v<p}$, Constante: ${v}={p}$')

  plt.plot(t,gf1(t,p+1),label=f'$y={yc}$,  $~(m ={p+1})$')
  plt.plot(t,gf1(t,p-1),label=f'$y={yd}$,  $~(m ={p-1})$')
  plt.plot(t,gf1(t,p),label=f'$y={yco}$,  $~(m ={p}$)')

  plt.grid()
  plt.legend()

  return pol

m,x = symbols('m x')

t = np.linspace(-9,9)
y = m*(x-1)+3-x
variacao(y,x,m,t)
```
